## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#begin by creating individual pulls for each of the drug regimens
capo_loc = clean_study_df.loc[clean_study_df['Drug Regimen'] == 'Capomulin']
rami_loc = clean_study_df.loc[clean_study_df['Drug Regimen'] == 'Ramicane']
infu_loc = clean_study_df.loc[clean_study_df['Drug Regimen'] == 'Infubinol']
ceft_loc = clean_study_df.loc[clean_study_df['Drug Regimen'] == 'Ceftamin']
#called each _loc to check df
```

```python
#get the last (greatest) timepoint for each mouse
max_capo = capo_loc.groupby(['Mouse ID'])['Timepoint'].max()
max_rami = rami_loc.groupby(['Mouse ID'])['Timepoint'].max()
max_infu = infu_loc.groupby(['Mouse ID'])['Timepoint'].max()
max_ceft = ceft_loc.groupby(['Mouse ID'])['Timepoint'].max()

# Merge these groups with the original dataframe to get the tumor volume at the last timepoint for each
#choose merge that only returns values with max timepoint, then display each to confirm
capo_merge = pd.merge(max_capo, clean_study_df, how="inner", left_on=["Mouse ID", "Timepoint"],
                                                                      right_on=["Mouse ID", "Timepoint"])

rami_merge = pd.merge(max_rami, clean_study_df, how="inner", left_on=["Mouse ID", "Timepoint"],
                                                                      right_on=["Mouse ID", "Timepoint"])

infu_merge = pd.merge(max_infu, clean_study_df, how="inner", left_on=["Mouse ID", "Timepoint"],
                                                                      right_on=["Mouse ID", "Timepoint"])

ceft_merge = pd.merge(max_ceft, clean_study_df, how="inner", left_on=["Mouse ID", "Timepoint"],
                                                                      right_on=["Mouse ID", "Timepoint"])

#capo_merge.head()
#rami_merge.head()
infu_merge.head()
#ceft_merge.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = [capo_merge, rami_merge, infu_merge, ceft_merge]

# Create empty list to fill with tumor vol data (for plotting)
tum_vol = []
drug_names = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
n=0

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
for x in drug_list:
    drug = x.loc[:,'Tumor Volume (mm3)']
    tum_vol.append(drug)
    # add subset 
    
    #determine variables for the quartiles
    quartiles = drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine outliers using upper and lower bounds
    lower = lowerq - (1.5*iqr)
    upper = upperq + (1.5*iqr)
    print(drug_names[n])
    n+=1
    print(f"Values found below {lower} and above {upper} may be outliers, and should be reviewed!")
          
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
outliers = dict(markerfacecolor='g', marker='*')
figl, axl = plt.subplots()

#set title, and labels
axl.set_title('Final Tumor Vol / Four Regimens')
axl.set_ylabel('Ending Tumor Volume')
axl.set_xlabel('Drug Regimen')

#call boxplot
axl.boxplot(tum_vol, labels=drug_names)

#save image and show plot
plt.savefig("Resources/tumor_boxplot.png")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#create a df for one mouse treated with Capomulin - chose g316
cap_g316 = capo_loc.loc[capo_loc['Mouse ID'] == 'g316']
cap_g316 #call data to check

#call plot and set labels, as well as title
plt.plot(cap_g316['Timepoint'],cap_g316['Tumor Volume (mm3)'], marker = "D", color = "blue")
plt.xlabel('Timepoint for Treatment')
plt.ylabel('Tumor Vol.')
plt.title('Tumor Vol vs Timepoint (Capomulin)- Mouse g316')


#save image and show plot
plt.savefig("Resources/tum_time_line.png")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
#calculate a groupby on the capo.merge by mouse and tumor and get the average
ave_tum_vol = capo_merge.groupby(['Mouse ID'])['Tumor Volume (mm3)'].mean()
ave_tum_vol #call data to check

#set variable for the mouse weight
m_weight = capo_merge['Weight (g)']
m_weight #call data to check

#set the variables, the labels, and title while calling the plot
x_values = m_weight
y_values = ave_tum_vol
plt.scatter(x_values, y_values, color = 'green')
plt.title('Tumor Volume vs Mouse Weight')
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')

#save image and show plot
plt.savefig("Resources/scatter_tumor.png")
plt.show()
```

## Correlation and Regression

```python
#Calculate the correlation coefficient and linear regression model for mouse weight and 
#average tumor volume for the Capomulin regimen

x_values = m_weight
y_values = ave_tum_vol

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values, color = 'green')
plt.plot(x_values, regress_values, 'r-')
plt.title('Tumor Volume vs Mouse Weight')
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.annotate(line_eq,(20,25),fontsize=15, color='red')

#save image and show plot
plt.savefig("Resources/scatter_regress.png")
plt.show()
```
